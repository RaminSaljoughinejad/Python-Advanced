```python
import qrcode
from PIL import Image
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.colormasks import RadialGradiantColorMask
import matplotlib. pyplot as plt
```

```python
img_name_path = "GitHub-Mark-64px.png"
img = Image.open(img_name_path)
```

```python
qr = qrcode.QRCode(
    version=4,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=20,
    border=4,
)
```

```python
text = input("Please Enter you text: ")

qr.add_data(text)
qr.make(fit=True)
```

```python
img = qr.make_image(image_factory=StyledPilImage,
                              embeded_image_path=img_name_path,
                              fill_color="red",
                              color_mask=RadialGradiantColorMask(back_color=(255,255,255),
                              center_color=(0,255,0),
                              edge_color=(0,0,255)))
```

```python
plt.imshow(img)
plt.show()
```

```python
img.save("GitHub.png")
print("Finish")
```
