```python
import qrcode
import matplotlib.pyplot as plt
```

```python
qr = qrcode.QRCode(
    version=4,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=20,
    border=4,
)
```

```python
text = input("Please Enter you text: ")

qr.add_data(text)
qr.make(fit=True)

img = qr.make_image()
```

```python
plt.imshow(img)
plt.show()
```

```python
img.save("youtube.png")
print("Finish")
```
